## üìä Strategic Data Sampling Theory

### ü§î Why Sample Large Datasets?

**Original Problem**:
- **20,640 samples** = computationally expensive
- **Decision trees** can become extremely complex
- **Training time** increases exponentially with data size
- **Memory usage** can become prohibitive

### üéØ Sampling Strategy: 40% Random Sample

**`df.sample(frac=0.40)`** explanation:
- **`frac=0.40`**: Take 40% of original data (8,256 samples)
- **Random sampling**: Maintains data distribution
- **No bias**: Every sample has equal probability of selection
- **Reproducible**: Set random_state for consistent results

### ‚öñÔ∏è Trade-offs Analysis:

#### **Benefits of Sampling**:
1. **Faster training**: 60% reduction in computation time
2. **Less memory**: Easier to handle on standard computers
3. **Simpler trees**: Reduces overfitting risk
4. **Faster experimentation**: Quick iteration during development

#### **Potential Risks**:
1. **Information loss**: Might miss rare patterns
2. **Reduced accuracy**: Less data = potentially worse performance
3. **Representation bias**: Small chance of skewed sample

### üìà When to Sample:
- **Large datasets** (>10,000 samples)
- **Initial exploration** and prototyping
- **Limited computational resources**
- **Quick model iteration** needed

### üö´ When NOT to Sample:
- **Small datasets** (<1,000 samples)
- **Final production models**
- **Rare event detection** (need all examples)
- **When computational resources are abundant**

### üí° Professional Practice:
1. **Start with sampling** for quick exploration
2. **Build final models** on full dataset
3. **Compare performance** between sampled and full data
4. **Document sampling decisions** for reproducibility

### üîß Alternative Sampling Methods:
- **Stratified sampling**: Maintain class proportions
- **Systematic sampling**: Every nth sample
- **Cluster sampling**: Sample groups, not individuals

## üî¢ Understanding Data Sampling (Why Less Can Be More)

### ü§î What Does `df.sample(frac=0.40)` Mean?

**Simple explanation**: Take a random 40% of all the houses, like picking names from a hat.

**Real-world analogy**: 
Imagine you want to know the average height of students in your school:
- **Option 1**: Measure ALL 1,000 students (takes forever)
- **Option 2**: Randomly pick 400 students and measure them (much faster, still accurate)

### üìä The Numbers:

#### **Before Sampling**:
- **20,640 houses** = A LOT of data
- **Takes long time** to process
- **Uses lots of computer memory**
- **Might crash your computer** if it's not powerful enough

#### **After Sampling (40%)**:
- **8,256 houses** = Much more manageable
- **Faster processing** (less than half the time)
- **Less memory usage** (fits on any computer)
- **Still representative** of all California houses

### üéØ Why 40%? (The Sweet Spot)

#### **Too Little Data (10%)**:
- ‚ùå Might miss important patterns
- ‚ùå Results might not be accurate
- ‚ùå Could be biased or unrepresentative

#### **Too Much Data (80%)**:
- ‚ùå Still slow to process
- ‚ùå Doesn't solve the computational problem
- ‚ùå Overkill for learning and experimentation

#### **Just Right (40%)**:
- ‚úÖ Fast enough for quick experiments
- ‚úÖ Large enough to find real patterns
- ‚úÖ Good balance of speed and accuracy
- ‚úÖ Perfect for learning and development

### üîç How Random Sampling Works:

```python
df.sample(frac=0.40)
```

**Step-by-step process**:
1. **Python looks at all 20,640 houses**
2. **Randomly assigns each house a number**
3. **Picks the first 8,256 houses (40%)**
4. **Returns a new DataFrame with just those houses**

**Key point**: Every house has an equal chance of being selected!

### üé≤ Why Random is Important:

#### **Good (Random) Sampling**:
- Expensive houses: 40% selected
- Cheap houses: 40% selected  
- Big houses: 40% selected
- Small houses: 40% selected
- **Result**: Sample represents the whole dataset

#### **Bad (Non-Random) Sampling**:
- Taking only the first 8,256 houses
- Could be biased (maybe all from one city)
- Results wouldn't apply to all of California

### üí° Professional Development Strategy:

1. **Start with a sample** (like 40%) for quick experimentation
2. **Test different algorithms** and parameters
3. **Once you find the best approach**, train on the full dataset
4. **Compare results** between sample and full data

### üîß Other Sampling Options:

```python
df.sample(n=5000)          # Take exactly 5,000 houses
df.sample(frac=0.25)       # Take 25% of houses  
df.sample(frac=0.60)       # Take 60% of houses

```

**Remember**: Sampling is like taste-testing soup - you don't need to eat the whole pot to know if it's good!

## üéØ Separating Features from Target (X and y Explained Simply)

### ü§î What are X and y in Machine Learning?

**Simple analogy**: Think of learning to drive a car

**X (Features)** = Everything you observe:
- Speed of other cars
- Weather conditions  
- Road signs
- Traffic lights

**y (Target)** = What you decide to do:
- Press the gas pedal
- Press the brake
- Turn left or right

### üè† In Our Housing Example:

#### **X (Features)** - House Characteristics:
All the information ABOUT the house that might affect its price:
- Location (latitude, longitude)
- Size (average rooms, bedrooms)
- Age of the house
- Neighborhood (income, population)

#### **y (Target)** - What We Want to Predict:
- **House Price** - The thing we're trying to guess

### üîç Understanding the Code:

```python
X = df.iloc[:,:-1]    # All columns EXCEPT the last one
y = df.iloc[:,-1]     # ONLY the last column
```

#### **Breaking Down `df.iloc[:,:-1]`**:
- `df.iloc`: "Access data by position numbers"
- `:`: "Take all rows"
- `:-1`: "Take all columns except the last one"
- **Result**: All house features (8 columns)

#### **Breaking Down `df.iloc[:,-1]`**:
- `df.iloc`: "Access data by position numbers"  
- `:`: "Take all rows"
- `-1`: "Take only the last column"
- **Result**: Just the prices (1 column)

### üìä Visual Representation:

```
Original DataFrame:
| MedInc | HouseAge | AveRooms | ... | Longitude | Price |
|--------|----------|----------|-----|-----------|-------|
| 8.32   | 41.0     | 6.98     | ... | -122.23   | 4.526 |

After Separation:

X (Features):                           y (Target):
| MedInc | HouseAge | AveRooms | ... |     | Price |
|--------|----------|----------|-----|     |-------|
| 8.32   | 41.0     | 6.98     | ... |     | 4.526 |
```

### üéØ Why Separate X and y?

#### **Machine Learning Process**:
1. **Show the computer X**: "Here are house features"
2. **Show the computer y**: "Here are the corresponding prices"  
3. **Computer learns**: "Oh, bigger houses cost more!"
4. **Test the computer**: "Given these new house features (X), what's the price (y)?"

#### **Real-World Analogy**:
Teaching a child to recognize animals:
- **X**: Show pictures of animals (features)
- **y**: Tell them what animal it is (target)
- **Learning**: Child learns to connect features to animal names
- **Testing**: Show new picture, child guesses the animal

### üí° Key ML Concepts:

#### **Features (X)** - The Input:
- Also called: predictors, independent variables, input variables
- **Purpose**: Information used to make predictions
- **Question they answer**: "What do we know about this house?"

#### **Target (y)** - The Output:
- Also called: response, dependent variable, output variable, label
- **Purpose**: What we're trying to predict
- **Question it answers**: "What's the price of this house?"

### üîß Common Mistakes Beginners Make:

‚ùå **Wrong**: Including price in X
- Computer would cheat (already knows the answer)

‚ùå **Wrong**: Having multiple targets in y  
- Computer gets confused about what to predict

‚úÖ **Correct**: Clear separation
- X = everything that helps predict
- y = the one thing we want to predict

### üé≤ Alternative Ways to Separate:

```python
# Method 1: By position (what we used)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# Method 2: By column name
X = df.drop('Price', axis=1)  # Remove price column
y = df['Price']               # Keep only price column

# Method 3: Explicit selection
X = df[['MedInc', 'HouseAge', 'AveRooms', ...]]

```

**Remember**: X = "What we know", y = "What we want to predict"

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
````

````python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model
````

````python
model.fit(X_train,y_train)
````

````python
y_pred=model.predict(X_test)
````

````python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
````

````python
#hyperparameter tuning
parameter={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
    'max_features':['auto','sqrt','log2']
}
regressor=DecisionTreeRegressor()
````

````python
from sklearn.model_selection import GridSearchCV
model=GridSearchCV(regressor,param_grid=parameter,cv=3,scoring='neg_mean_squared_error')
````

````python
model.fit(X_train,y_train)
````

````python
model.best_params_
````

````python
final_model=DecisionTreeRegressor(criterion='squared_error',max_depth=8,max_features='log2',splitter='best')
````

````python
final_model.fit(X_train,y_train)
````

````python
from sklearn import tree
plt.Figure(figsize=(12,10))
tree.plot_tree(final_model,filled=True)
````

````python
y_pred=model.predict(X_test)
````
